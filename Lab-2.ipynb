# MLP - Multi Layer Perceptron

### Load Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

###  Define Activation Function

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

### derivative of backpropagation

```python
def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)
```

### intialise dataset

- y = 2x

```python
x = np.array([1, 2, 3, 4, 5])
labels = 2*x
```

### intialise weight and bias   ( Dono ke weight intialise kiye)

```python
weights_input_hidden = np.random.rand()
weights_hidden_output = np.random.rand()
bias_hidden = np.random.rand()
bias_output = np.random.rand()
```

## For each Epoch 

- compute hidden layer input
- apply activation function
- compute output layer input
- apply activation fun to obtain prediction

## Backpropagation
- compute error at output layer
- compute gradient of loss with respect to output weights
- Propagate error to hidden layer
- Compute gradient for hidden layer weight

```python
for epoch in range(10000):
    z1 = np.dot(x, weights_input_hidden) + bias_hidden
    a1 = sigmoid(z1)
    z2 = np.dot(a1, weights_hidden_output) + bias_output
    y = sigmoid(z2)

    mse = np.mean((labels - y) ** 2)
    print(f'Epoch {epoch+1}, Loss: {mse}')

    error = y - labels
    d_output = error * sigmoid_derivative(z2)
    error_hidden = d_output * weights_hidden_output
    d_hidden = error_hidden * sigmoid_derivative(z1)
    weights_hidden_output -= np.dot(a1, d_output) * 0.01
    weights_input_hidden -= np.dot(x, d_hidden) * 0.01
    bias_output -= np.sum(d_output, axis=0) * 0.01
    bias_hidden -= np.sum(d_hidden, axis=0) * 0.01

print("Final predictions:", y)
```
